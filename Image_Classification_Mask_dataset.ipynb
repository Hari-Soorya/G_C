```python
import numpy as np
import pandas as pd
import os
from skimage.transform import resize
from skimage.io import imread
```

```python
flat_arr=[]
target_arr=[]
Category=['with_mask1','without_mask1']
```

```python
datadir='/content/drive/MyDrive/Mask_datatest'
for i in Category:
  print('Loading ',i)
  path=os.path.join(datadir,i)
  for img in os.listdir(path):
    img_arr=imread(os.path.join(path,img))
    img_resiz=resize(img_arr,(150,150,3))
    flat_arr.append(img_resiz.flatten())
    target_arr.append(Category.index(i))
  print('Loaded ',i)
```

```python
df=pd.DataFrame(flat_arr)
df
```

```python
df['target']=target_arr
df
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.svm import SVC
sv=SVC()
sv.fit(x_train,y_train)
y_pre=sv.predict(x_test)
y_pre
```

```python
from sklearn.metrics import accuracy_score,classification_report,ConfusionMatrixDisplay
print(accuracy_score(y_test,y_pre))
```

```python
print(classification_report(y_test,y_pre))
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pre))
```
